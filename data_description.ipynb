```python
# index : 데이터를 집계(그룹화)하는 기준
    # 가족관계를 기준으로 데이터를 집계하고,
    # 다시 workclass를 기준으로 데이터를 집계함
# columns : index로 집계한 그룹을 다시 columns을 기준으로 그룹화함
# value : index와 columns으로 그룹화한 그룹 중에서 출력하고 싶은 데이터
# aggfunc : value에 적용하고 싶은 집계 연산

pd.pivot_table(df, values='hoursperweek', index=['relationship', 'workclass'],
               columns=['label'], aggfunc=np.mean).round(2)
```

```python
# 두 변수 간의 관계에서 빈도를 확인하는 방법
# 함수에 두 개의 변수를 전달하면, 이 두 변수의 모든 쌍별 조합의 빈도를 얻을 수 있음

pd.crosstab(df['label'], df['relationship'])
```

```python
# crosstab에 normalize를 적용함

pd.crosstab(df['label'], df['relationship'], normalize=True)
```

```python
# long form 데이터 만들기
    # long form 데이터 : 각 행의 관찰값이 잠재적 column이 될 수 있는 형식
    # variable이라는 이름의 열에 각 개별 관찰값이 기록되기 때문에 표가 '길어질 수 밖에 없음'

import pandas.util.testing as tm; tm.N = 3

def unpivot(frame):
    N, K = frame.shape
    data = {'value' : frame.values.ravel('F'),
            'variable' : np.asarray(frame.columns).repeat(N),
            'date' : np.tile(np.asarray(frame.index), K)}
    return pd.DataFrame(data, columns=['date', 'variable', 'value'])

df = unpivot(tm.makeTimeDataFrame())
df
```

```python
# long form 데이터에서 wide format으로 변환
    # variable의 각 관측값이 column으로 사용됨

df_pivot = df.pivot(index='date', columns='variable', values='value')
df_pivot
```
